## Introduce the sigmoid function:

## $$ f(y) = \frac{1}{1 + e^{-a y}}$$

## Breakout Exercise: 
- ## Write a function, sigmoid(y, a=1).
- ## Plot the sigmoid function, with $a$ equal to its default value of 1.  Ask yourself, does this function behave according to your expectation?
- ## Plot the simoid function with 5 different $a$'s, [1, 3, 5, 7, 9]

```python
def sigmoid_fun (y, a=1):
    return 1 / (1+np.exp(-(y*a)))
```

```python
y = np.linspace(-10, 10, 100)
plt.plot(y, sigmoid_fun(y))
plt.title('Sig Function')
plt.show()
```

```python
a=[1, 3, 5, 7, 9]
for _ in a:
    plt.plot(y, sigmoid_fun(y, _), label='a = {:.1f}'.format(_))

plt.title('Sig Function')
plt.legend()
plt.show()
```

```python
y = np.linspace(-5, 5, 1000)
sinh = (np.exp(y) - np.exp(-y))/2
cosh = (np.exp(y) + np.exp(-y))/2
z = sinh/cosh
zprime = 4 / (np.exp(y) + np.exp(-y))**2
derivativez = 1 - z**2
plt.plot(y, derivativez)
plt.show()
```

```python
epsilon = (z[1]-z[0]) / (y[1]-y[0])
two_eps = 2 * epsilon
z_der = np.array([])
for yi in y:
    z_pr = (np.tanh(yi + epsilon) - np.tanh(yi - epsilon)) /two_eps
    z_der = np.append(z_der, z_pr)
    
plt.plot(y, z_der, '.', label='Numerical Derivative')
plt.plot(y, 1 - z ** 2, label='Analytic Derivative')
plt.legend()
plt.show()

```

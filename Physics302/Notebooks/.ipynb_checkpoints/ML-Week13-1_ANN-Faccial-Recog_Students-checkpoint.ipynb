## The Confusion Matrix:  Evaluate the accuracy of a classification.

## Definition: a confusion matrix C is such that C\_{i, j} is equal to the number of observations known to be in group i but predicted to be in group j.

From: http://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html


## Each row should add up to the number of objects in each class.

```python
'''
3 dogs: all correctly classified

3 cats: 1 is incorrectly classified as dog, 2 classficied correctly

1 racoon: incorrectly classified as cat

'''

y_true = ["cat", "dog", "cat", "cat", "dog", "racoon", "dog"]
y_pred = ["dog", "dog", "cat", "cat", "dog", "cat", "dog"]
confusion_matrix(y_true, y_pred, labels=["dog", "racoon", "cat"])
```

## Part I: Apply it to one person -- say the first in the data set


```python
target_names = lfw_people.target_names
print(target_names)
```

```python
n_samples, h, w = lfw_people.images.shape

X = lfw_people.data
n_features = X.shape[1]

y = lfw_people.target
target_names = lfw_people.target_names

print('Fitting the classifier to the training set')
t0 = time.time()

clf = MLPClassifier(solver='lbfgs', alpha=1e-10, hidden_layer_sizes=(100,), activation = 'tanh', \
                    random_state=42, max_iter=10000, learning_rate_init=0.001)

X1 = np.atleast_2d(X[0])
y1 = np.atleast_2d(y[0])

clf = clf.fit(X1, y1)
print('Training time: {} sec'.format( time.time()-t0))

y_pred = clf.predict(X1)
print("y, y_pred", y[0], y_pred[0])
print("true name: {}, predicted name:{}".format(target_names[y[0]], target_names[y_pred[0]]))

plt.imshow(X1.reshape(50, -1), cmap='gray')
plt.axis('off')
plt.show()
print('Number of features', n_features)
```

```python
clf = clf.fit(X_train, y_train)
print('Training time: {} sec'.format( time.time()-t0))

y_pred = clf.predict(X_test)
y_pred_test = clf.predict_proba(X_test)

train_score = clf.score(X_train, y_train)
test_score = clf.score(X_test, y_test)

y_score = clf.predict_proba(X_test)

print('Train Score: ',train_score)
#try to get 0.7 for test_score
print('Test Score: ',test_score)

# print("y, y_pred", y[0], y_pred[0])
# print("true name: {}, predicted name:{}".format(target_names[y[0]], target_names[y_pred[0]]))

# plt.imshow(X1.reshape(50, -1), cmap='gray')
# plt.axis('off')
# plt.show()
# print('Number of features', n_features)
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
```

```python
#Binarize the output
print('y shape before binarizing: {}'.format(y.shape))
y_binary = label_binarize(y, classes=[0, 1, 2, 3, 4, 5, 6])
print('y shape after binarizing: {}'.format(y_binary.shape))
X_train, X_test, y_train_bin, y_test_bin = train_test_split(X, y_binary, test_size=0.25, random_state=42)

y_score_bin = clf.predict_proba(X_test)
```

```python
TPR = dict()
FPR = dict()

roc_auc = dict()
name = 'Colin Powell'
for i in range(n_classes):
    FPR[i], TPR[i], _ = roc_curve(y_test_bin[:,i], y_score_bin[:,i])
    roc_auc[i] = auc(FPR[i], TPR[i])
    if target_names[i]==name:
        plt.plot(FPR[i], TPR[i], label='AUC = {1:0.4f}'.format(name,roc_auc[i]))

plt.plot([0, 1], [0, 1], linestyle='--')
# plt.ylim(.6, 1)
# plt.xlim(0, .4)
plt.title('ROC Curve for: '+ name)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

```python
def sigmoid(x):
    '''The logistic function as the sigmoid'''

    return 1.0/(1.0 + np.exp(-x))

def sigmoid_pr(z):
    '''derivative of the logistic function'''
    return z * (1 - z)
```

```python
def simple_nn(X, w1, w2):
    #return z, b, A, a
#     a0=(w1[0][0] * X[0]) + ( w1[1][0] * X[1] )
#     a1=(w1[0][1] * X[0]) + ( w1[1][1] * X[1] )
#     a=[a0, a1]
    a = np.dot(X, w1)
    
#     A0=sigmoid(a0) 
#     A1=sigmoid(a1)
#     A=[A0, A1]
    A = sigmoid(a)
    
#     b=(A0*w2[0])+(A1*w2[1])
    b = np.dot(A, w2)
    
    z=sigmoid(b)
    
    return z, b, A, a
    
X = np.array([0.35, 0.9])   
w1 = np.array([[0.1000, 0.4000], [0.8000, 0.6000]])
print('w1:\n', w1)
w2 = np.array([0.3000, 0.9000])
print('w2:\n', w2)

z, b, A, a = simple_nn(X, w1, w2)
print('a: ', a)
print('A: ', A)
print('b: ', b)
print('output (z) of nn: ', z)
```

```python
def training_nn(X, y, z, w1, w2):
    alpha=1
    error = y-z
    delta2 = error * sigmoid_pr(z)
    delta_w2 = np.dot(delta2, A)
#     print('delta2: ', delta2)
#     print('delta_w2: ', delta_w2)
    w2 += (alpha * delta_w2)
#     print('w2: ', w2)
#     print('\n')
    
#     print('sigmoid_pr(A): ', (A))
#     delta1_0 = delta2 * w2[0] * A[0]*(1-A[0])
#     delta1_1 = delta2 * w2[1] * A[1]*(1-A[1])
#     delta1 = np.atleast_2d(np.array([delta1_0, delta1_1]))
    delta1 = np.atleast_2d(delta2 * sigmoid_pr(A) *w2)
    X = np.atleast_2d(X)
    delta_w1 = np.dot(X.T, delta1)
#     print('delta1: ', delta1)
#     print('delta_w1: \n', delta_w1)
    w1 += (alpha * delta_w1)
    #print('w1: \n', w1)
    
    return w2, w1

y=0.5
X = np.array([0.35, 0.9]) 

w1 = np.array([[0.1000, 0.4000], [0.8000, 0.6000]])
print('w1:\n', w1)
w2 = np.array([0.3000, 0.9000])
print('w2:\n', w2)


w2new, w1new = training_nn(X, y, z, w1, w2)
print('w1new:\n', w1new)
print('w2new:\n', w2new)
```

```python
#Solving the XOR
def simple_nn(X, w1, w2):
    a = np.dot(X, w1)
    A = sigmoid(a)
    b = np.dot(A, w2)
    z=sigmoid(b)
    return z, b, A, a
    
X = np.array([0.35, 0.9])   
w1 = np.array([[0.1000, 0.4000], [0.8000, 0.6000]])
print('w1:\n', w1)
w2 = np.array([0.3000, 0.9000])
print('w2:\n', w2)

z, b, A, a = simple_nn(X, w1, w2)
print('a: ', a)
print('A: ', A)
print('b: ', b)
print('output (z) of nn: ', z)
```

```python
def training_nn(X, y, z, w1, w2):
    alpha=1
    error = y-z
    delta2 = error * sigmoid_pr(z)
    delta_w2 = np.dot(delta2, A)
#     print('delta2: ', delta2)
#     print('delta_w2: ', delta_w2)
    w2 += (alpha * delta_w2)
#     print('w2: ', w2)
#     print('\n')
    
#     print('sigmoid_pr(A): ', (A))
#     delta1_0 = delta2 * w2[0] * A[0]*(1-A[0])
#     delta1_1 = delta2 * w2[1] * A[1]*(1-A[1])
#     delta1 = np.atleast_2d(np.array([delta1_0, delta1_1]))
    delta1 = np.atleast_2d(delta2 * sigmoid_pr(A) *w2)
    X = np.atleast_2d(X)
    delta_w1 = np.dot(X.T, delta1)
#     print('delta1: ', delta1)
#     print('delta_w1: \n', delta_w1)
    w1 += (alpha * delta_w1)
    #print('w1: \n', w1)
    
    return w2, w1

y=0.5
X = np.array([0.35, 0.9]) 

w1 = np.array([[0.1000, 0.4000], [0.8000, 0.6000]])
print('w1:\n', w1)
w2 = np.array([0.3000, 0.9000])
print('w2:\n', w2)


w2new, w1new = training_nn(X, y, z, w1, w2)
print('w1new:\n', w1new)
print('w2new:\n', w2new)
```
